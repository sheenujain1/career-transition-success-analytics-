```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/user/Downloads/archive (9)/career_change_prediction_dataset.csv")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_')
print("Cleaned Columns:\n", df.columns.tolist())
```

```python
print(df['current_occupation'].unique())
```

```python
tech_keywords = ['Software Developer', 'Business Analyst', 'Mechanical Engineer']
df['is_tech_designation'] =  df['current_occupation'].str.lower().isin(tech_keywords).map({True: 'Tech', False: 'Non-Tech'})
```

```python
df['field_of_study'] = df['field_of_study'].str.strip()
df['current_occupation'] = df['current_occupation'].str.strip()
```

```python
def bin_score(score):
    if score <=4:
        return 'Low'
    elif score <=7:
        return 'Medium'
    else:
        return 'High'
    
df['satisfaction_tier'] = df['job_satisfaction'].apply(bin_score)
df['wlb_tier'] = df['work_life_balance'].apply(bin_score)
df['security_tier'] = df['job_security'].apply(bin_score)
df['sg_tier'] = df['skills_gap'].apply(bin_score)
df['networks_tier'] = df['professional_networks'].apply(bin_score)
```

```python
df['switch_intent'] = df['likely_to_change_occupation'].map({1: 'Likely to Pivot', 0: 'Likely to Stay'})
df['has_mentorship'] = df['mentorship_available'].map({1: 'Yes', 0: 'No'})
df['has_freelance_exp'] = df['freelancing_experience'].map({1: 'Yes', 0: 'No'})

df.drop(columns=['likely_to_change_occupation', 'mentorship_available'], inplace=True)
```

```python
print("Missing values per column:\n", df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
invalid_exp_mask = df['years_of_experience'] > (df['age'] - 15)
print(f"Found {invalid_exp_mask.sum()} rows with unrealistic experience metrics.")
```

```python
df.loc[invalid_exp_mask, 'years_of_experience'] = df['age'] - 18
```

```python
df.to_csv("dashboard_ready_career_data.csv", index=False)
print("Data is pristine and saved as 'dashboard_ready_career_data.csv'!")
```
